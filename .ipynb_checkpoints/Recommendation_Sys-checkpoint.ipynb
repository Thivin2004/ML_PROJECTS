```python
credits=pd.read_csv(r"E:\ML_PROJECTS\Movie_Recommendation_System\dataset\tmdb_5000_credits.csv")
movies=pd.read_csv(r"E:\ML_PROJECTS\Movie_Recommendation_System\dataset\tmdb_5000_movies.csv")
```

```python
movies=movies.merge(credits,on ='title')
```

```python
movies
```

```python
movies=movies[['movie_id','title','cast','crew','genres','keywords','overview']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.iloc[0]['genres']
```

```python
import ast
def convert(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0]['keywords']
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.iloc[0]['keywords']
```

```python
movies.head()
```

```python
def convert_cast(c):
    l=[]
    counter=0
    for i in ast.literal_eval(c):
        if counter<3:
            l.append(i['name'])
            counter+=1
    return l
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
movies.iloc[0]['crew']
```

```python
def fetch_dir(direc):
    l=[]
    for i in ast.literal_eval(direc):
        if i['job'] == 'Director':
            l.append(i['name'])
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_dir)
```

```python
movies.head()
```

```python
movies.iloc[0]['overview']
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies.iloc[0]['overview']
```

```python
def remove_space(space):
    l=[]
    for i in space:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['cast']=movies['cast'].apply(remove_space)
movies['crew']=movies['crew'].apply(remove_space)
movies['genres']=movies['genres'].apply(remove_space)
movies['keywords']=movies['keywords'].apply(remove_space)
```

```python
movies.head()
```

```python
movies['tags']= movies['cast'] + movies['crew'] + movies['genres'] + movies['keywords']  + movies['overview']
```

```python
movies.head()
```

```python
movies.iloc[0]['tags']
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df.iloc[0]['tags']
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: [tag.lower() for tag in x])
```

```python
new_df.head()
```

```python
new_df.iloc[0]['tags']
```

```python
import  nltk
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python

def stem(text):
    if isinstance(text, list):  # Convert list to string
        text = " ".join(text)
    return " ".join(ps.stem(i) for i in text.split())
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vector[0]
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import  cosine_similarity
similar=cosine_similarity(vector)
```

```python
similar
```

```python
similar.shape
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similar[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
